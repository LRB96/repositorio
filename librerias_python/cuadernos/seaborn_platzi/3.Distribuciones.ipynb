```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Seaborn ofrece algunas gráficas u opciones que te permiten trabajar con distribuciones orientadas a tipos de datos numéricos. Veamos como puedes lograrlo utilizando código.

Uso de .kde y .displot para graficar distribuciones

Gráfico para distribuciones y que se acumulen cumulative

sns.histplot(data = tips, x= 'tip', bins = 15, cumulative=True)
sns.histplot(data = tips, x= 'tip', bins = 15, hue= 'sex', stat = 'density', multiple= 'dodge');

Uso de .kde
Gráfica de densidad

sns.kdeplot(data=tips, x='tip', hue='sex'); 

sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True);

Crea una área debajo de la curva
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True);

Uso de .displot
Gráfica de distribución

Cambiar tipo de gráfica con kind.
sns.displot(data = tips, x = 'tip', hue = 'sex',kind= 'hist', multiple= 'stack');

Cargar un dataset de seaborn Guarda la grafica como imagen para descargar plt.savefig(“name.png”)

bins=numero establece el numero de barras que tiene la grafica cumulative=True acumula los valores, y permite ver el comportamiento acumulativo de las frecuencias en el histograma Con multiple="" establezco como quiero presentar mis datos

bw_adjust=1 ajusta la gráfica, y los parámetros que estamos manejando en los ejes. kdeplot grafica de densidad shade=True crea un area bajo la curva para que sea mas facil de ver bw_adjust=1 ajusta directamente la grafica, y los parametros que se manejan en los ejes

ecdplot crea un gracfico escalonado.

Para usar graficas de distribucion se usa displot como herramienta general, y en kind="" se agrega el tipo, ejemplo, kde o hist

Para graficas de distribuciones displot(kind="") Para graficas de variables categoricas catplot(kind="")

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips,x="tip",bins=15,cumulative=False,hue="sex",stat="probability");
```

```python
sns.histplot(data=tips,x="tip",bins=15,cumulative=False,hue="sex",stat="percent");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",bins=15,cumulative=False,stat="density");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",cumulative=False,stat="count");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",cumulative=False,stat="probability",multiple="stack");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",cumulative=False,stat="probability",multiple="layer");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",cumulative=False,stat="percent",multiple="dodge");
```

```python
sns.histplot(data=tips,x="tip",hue="sex",cumulative=False,stat="count",multiple="fill");
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=True,shade=True,bw_adjust=1); 
```

El grafico ecdf es escalonado, lo que permite ver la proporcion de la variables en ella.

```python
sns.ecdfplot(data=tips,x="tip",hue="sex",stat="count"); 
```

```python
sns.displot(data=tips,x="tip",hue="sex",kind="hist",multiple="stack")
plt.savefig("figura1.png",dpi=300);
```
