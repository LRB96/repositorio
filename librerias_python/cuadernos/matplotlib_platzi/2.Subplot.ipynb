```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

Cómo acceder a la matriz de gráficos en Subplot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
Parámetros en Subplot
Los parámetros de subplot funcionan:

Filas	Columnas	Index
Crear fila y dos columnas
Crear una matriz de gráficos de una fila y dos columnas
plt.subplot(1,2,1) # Grafico 1
plt.plot(x,y, 'r--') 

plt.subplot(1,2,2)  # Gráfico 2
plt.pie(y)
plt.show()

Agregar diferentes Plot al gráfico

plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()

Invertir la matriz de los gráficos

plt.subplot(2,1,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,1,2) #Grafico 2
plt.pie(y)         #Plot 1  
plt.show()

Subplot Platzi

```python
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y);
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.plot(y,x,"b:")
plt.subplot(1,2,2)
plt.pie(y);
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"r--")
plt.plot(y,x,"b:")
plt.subplot(2,1,2)
plt.pie(y);
```

```python
#pie, boxplot e hist solo llevan un parámetro
#plot,plot,pie,boxplot,bar,hist

x= [ 1,2,3,4,5,7,9,14]
y= [ 1,4,8,6,5,10,9,11]

plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.plot(x,y,marker="o")
plt.subplot(2,3,3)
plt.bar(x,y)
plt.subplot(2,3,4)
plt.pie(x)
plt.subplot(2,3,5)
plt.boxplot(x)
plt.subplot(2,3,6)
plt.hist(x);
```
