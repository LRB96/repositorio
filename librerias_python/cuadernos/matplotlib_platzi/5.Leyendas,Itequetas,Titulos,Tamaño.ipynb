```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.

Cómo dar contexto a los gráficos

Creamos nuestras gráficas comunes

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()

Contexto a los gráficos con títulos

Para crear los títulos realiza lo siguiente:

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title("Relacion X - Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")


Contexto agregando etiquetas

Agrega labels en los ejes

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', )
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")

Cambiar el tamaño a las figuras

Realiza la siguiente acción:

fig, (ax1,ax2) = plt.subplots(1,2, figsize= (1,10))

Aplicar etiquetas (por defecto pone la mejor opción)

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()

* Modificar la localización de la etiqueta

ax1.legend(loc = "upper center")

Otras ubicaciones

upper right	upper left	lower left	lower right	right

```python
x=np.linspace(0,5,11)
y=np.sin(x)
print(x)
print(y);
```

```python
#Para que muestre el legend se debe poner label="" al momento de crear la grafica.
#upper right	upper left	lower left	lower right	right--upper center
#bbox_to_anchor junto con loc="" en legend sirve ubicar la leyenda de la grafica.
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(8,8))
ax1.plot(x,y,"r",label="Relacion X - Y")
ax1.set(xlabel="X",ylabel="Y",title="Relacion X - Y")
ax1.legend(loc="lower left",bbox_to_anchor=(0.1,0.1))

ax2.plot(y,x,"b",label="Relacion Y- X")
ax2.set(xlabel="X",ylabel="Y",title="Relacion Y - X")
ax2.legend(loc="lower right",bbox_to_anchor=(0.6,0.2));
```

```python
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y,label="X")
ax[1].plot(y,x,label="Y")
ax[0].set(xlabel="X",ylabel="Y",title="Relación X - Y")
ax[1].set(xlabel="Y",ylabel="X",title="Relación Y - X")
ax[0].legend(loc="lower left")
ax[1].legend(loc="lower left");
#upper right	upper left	lower left	lower right	right--upper center
```
