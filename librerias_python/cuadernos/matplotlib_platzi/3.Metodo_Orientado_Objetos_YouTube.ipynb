Hay distintas maneras de hacer gráficas dentro de Matplotlib, ya vimos pyplot; sin embargo, es muy complicado personalizarla y entrar a ciertos parámetros.

Método orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes grá ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.

Es un poco más complicado, pero en el mismo gráfico podemos personalizarlo mucho mejor.

    Diferencias entre Pyplot y Object Oriented
    
Pyplot	Object Oriented
Rápido	Mayor personalización
Fácil	Más código

Una sola figura	Más amigable a múltiples diagramas

Implementar el método orientado a objetos para graficar en Matplotlib

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2

Parámetros de axes:
    
Pos. Eje x	Pos. Eje y	Size graph eje x	Size graph eje y

Utilizamos el concepto de objeto para crear una gráfica más personalizable

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'b')
plt.show()


Podemos crear varias gráficas dentro del lienzo llamando el método fig.

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()

```python
#Datos de ejemplo
dias=[1,2,3,4,5,6,7]
gasto=[10,25,15,30,35,5,20]
gasto2=[12,13,43,51,3,32,45]
gasto3=[21,42,45,64,1,8,24]
```

```python
#Un objeto define un figura.
#Esa figura es un lienzo, en el cual voy a meter gráficas.
#A cada una de esas gráficas se les llama axis.
#Estos axe a su vez pueden tener diferentes ejes.
#Esos ejes se van a llamar axis.

#Figure es el lienzo.
#Los axes son las gráficas
#Los axis son (x,y,z) de dichas gráficas.
```

```python
plt.plot(dias,gasto,color="g",marker="o",markersize=4,linewidth=7);
```

```python
fig,ax=plt.subplots()
ax.plot(dias,gasto)
ax.plot(dias,gasto2)
ax.plot(dias,gasto3)
ax.set(xlabel="Días",ylabel="Gasto",title="Gastos")
ax.legend([gasto,gasto2,gasto3]);


#ig,ax=plt.subplots()
#ax.plot(dias,gasto)
#ax.plot(dias,gasto2)
#ax.plot(dias,gasto3)
#ax.set(xlabel="Días",ylabel="Gasto",title="Gastos")
#ax.legend([gasto,gasto2,gasto3]);
```

```python
fig,ax=plt.subplots()
x=np.linspace(0,20,2000)
ax.plot(x,np.sin(x),'r',linestyle="dashed")
ax.plot(x,np.sin(x+3),'b',linestyle="dotted");
```

```python
#Graficar el resto de la matrix 1,1  2,0  2,1
fig,ax=plt.subplots(3,2,figsize=[9,12])
ax[0,0].scatter(dias,gasto,marker="o",color="r",linestyle="--",label="$Gasto1$")
ax[0,1].plot(dias,gasto2,marker="s",color="b",linestyle="-",label="Gasto2")
ax[1,0].plot(dias,gasto3,marker="D",color="m",linestyle="dotted",label="Gasto3")
ax[1,1].bar(dias,gasto,color="r",linestyle="--",label="Gastos4")
ax[2,0].pie(gasto2)
ax[2,1].hist(gasto3,color="m",linestyle="dotted",label="Gasto6")
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()
ax[2,1].legend();


#fig,ax=plt.subplots(3,2,figsize=[9,12]) #,sharey=False
#ax[0][0].plot(dias,gasto,marker="o",color="r",linestyle="--",label="Gastos1")
#ax[0][1].plot(dias,gasto2,marker="s",color="b",linestyle="-",label="Gastos2")
#ax[1][0].plot(dias,gasto3,marker="D",color="m",linestyle="dotted",label="Gastos3")
#ax[1][1].plot(dias,gasto,marker="o",color="r",linestyle="--",label="Gastos4")
#ax[2][0].plot(dias,gasto2,marker="D",color="y",linestyle="-",label="Gasto5")
#ax[2][1].plot(dias,gasto3,marker="s",color="m",linestyle="dotted",label="Gastos6")
#ax[0][1].set_xlabel("Días")
#ax[0][1].set_title("Gastos2")
#ax[1][0].legend([gasto]);
```
