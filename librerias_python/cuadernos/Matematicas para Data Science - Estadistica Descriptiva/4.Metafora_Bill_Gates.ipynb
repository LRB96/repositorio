```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./cars.csv",sep=",",header=0).rename(columns=str.lower)
df
```

```python
sns.displot(df,x="price_usd",hue="engine_type",multiple="dodge");
```

```python
Q7=df[(df["manufacturer_name"]=="Audi") & (df["model_name"]=="Q7")]
Q7.head()
```

```python
sns.displot(Q7,x="price_usd",hue="year_produced");
```

```python
Chrysler=df[(df["manufacturer_name"]=="Chrysler") & (df["model_name"]=="Voyager") & (df["year_produced"]>2001) & (df["engine_fuel"]=="diesel")]
Chrysler
```

```python
sns.displot(Chrysler,x="price_usd",hue="color")
plt.savefig("Chrysler_By_Color",dpi=300)
```
