    Funciones trascendentes
    
Son funciones que no pueden ser expresadas con polinomios.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
                        #Funciones trigonometricas

N=100
x=np.linspace(-10,10,N)
def f(x):
    return np.cos(x)
y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
                        #Funciones exponencial
N=100

def f(x):
    return np.exp(x)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

                        #Funcion logaritmo
    
El logaritmo está definido por la relación:

logb(x)=n⟺x=bn 

donde:

b  es la base.
n  es el exponente al que está elevado la base.
x  es el resultado de elevar la base  b  al exponente  n 
Ejemplo:

Teniendo b=2 y n=8, entonces:

28=256 

Por lo que  x=256 . Calculando el logaritmo base 2 de  x  es:

log2(256)=8

```python
N=100
x=np.linspace(0.001,256,N)
def f(x):
    return np.log2(x)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```
