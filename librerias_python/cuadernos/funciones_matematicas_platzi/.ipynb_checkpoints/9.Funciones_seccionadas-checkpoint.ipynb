```python
import matplotlib.pyplot as plt
import numpy as np
```

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside:

H(x)={0, para x<0
      1, para x≥0

```python
Una funcion seccionada se comporta diferente dependiendo de las funciones.
```

```python
#enumerate es una funcion que nos permite obtener el indice, y el elemento de un arreglo en python
    
def H(x):
    Y=np.zeros(len(x))
    for idx,x in enumerate(x):
        if x> 0:
            Y[idx]=1
    return Y
N=1000

x=np.linspace(-10,10,N)
y=H(x)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
N=100
x=np.linspace(-10,10,N)
def f(x):
    return np.int0(x>=0)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
##El valor absoluto de un numero se calcula tomando la raiz cuadrada de su cuadrado.
N=100
x=np.linspace(-10,10,N)

def f(x):
    return np.absolute(x)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```
