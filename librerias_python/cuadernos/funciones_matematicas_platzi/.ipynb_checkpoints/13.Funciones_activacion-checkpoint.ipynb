```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Funciones de activación == Darle un valor de salida diferente a todo el proceso del perceptrón
```

```python
N=100
x=np.linspace(-5,5,N)
```

```python
#Funcion lineal y=m*x+b
#Función lineal: mantener valores a lo largo de un proceso, predecir venta

def f(x):
    return x

fig,ax=plt.subplots()
ax.plot(x,f(x))
ax.grid();
```

```python
#Función escalón o de Heaviside
#Función escalón o de Heaviside: clasificaciones categóricas, ej. binarias

def H(x):
    return np.int0(x>=0)

y=H(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
#Funcion sigmoide
#Función sigmoide: regresión logística, para probabilidades

def f(x):
    return 1/(1+np.exp(-x))

y=f(x)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
#Función tangente hiperbólica: para escalamiento

def f(x):
    return np.tanh(x)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```

```python
#Función ReLU: simulación de “neuronas muertas”

def f(x):
    return np.maximum(0,x)

y=f(x)
fig,ax=plt.subplots()
ax.plot(x,y)
ax.grid();
```
