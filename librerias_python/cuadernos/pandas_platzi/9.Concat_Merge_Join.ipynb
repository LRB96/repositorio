Left join: Da prioridad al DataFrame de la izquierda. Trae siempre los datos de la izquierda y las filas en común con el DataFrame de la derecha.

Right join: Da prioridad al DataFrame de la derecha. Trae siempre los datos de la derecha y las filas en común con el DataFrame de la izquierda.

Inner join: Trae solamente aquellos datos que son común en ambos DataFrame

Outer join: Trae los datos tanto del DataFrame de la izquierda como el de la derecha, incluyendo los datos que comparten ambos.

                                            Concat

```python
df1=pd.DataFrame({"A":["A0","A1","A2","A3"],
                  "B":["B0","B1","B2","B3"],
                  "C":["C0","C1","C2","C3"],
                  "D":["D0","D1","D2","D3"]})
df1
```

```python
df2=pd.DataFrame({"A":["A4","A5","A6","A7"],
                  "B":["B4","B5","B6","B7"],
                  "C":["C4","C5","C6","C7"],
                  "D":["D4","D5","D6","D7"]})
df2
```

```python
#ignore_index=True sirve para que se creen indices nuevos.
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1,ignore_index=True)
```

                                                        Merge

```python
izq=pd.DataFrame({"key":["K0","K1","K2","K3"],
                  "A":["A0","A1","A2","A3"],
                  "B":["B0","B1","B2","B3"]})
izq
```

```python
der=pd.DataFrame({"key":["K0","K1","K2","K3"],
                 "C0":["C0","C1","C2","C3"],
                 "D0":["D0","D1","D2","D3"]})
der
```

```python
#Con on="columna_comun" se establecen los valores que ambos df tienen.
izq.merge(der,on="key")
```

```python
izq=pd.DataFrame({"key":["K0","K1","K2","K3"],
             "A":["A0","A1","A2","A3"],
             "B":["B0","B1","B2","B3"]})

der=pd.DataFrame({"key_2":["K0","K1","K2","K3"],
                 "C0":["D0","C1","C2","C3"],
                 "D0":["D0","D1","D2","D3"]})
```

```python
izq.merge(der,left_on="key",right_on="key_2")
```

```python
izq=pd.DataFrame({"key":["K0","K1","K2","K3"],
             "A":["A0","A1","A2","A3"],
             "B":["B0","B1","B2","B3"]})

der=pd.DataFrame({"key_2":["K0","K1","K2",np.nan],
                 "C0":["D0","C1","C2","C3"],
                 "D0":["D0","D1","D2","D3"]})
izq
```

```python
der
```

```python
izq.merge(der,left_on="key",right_on="key_2",how="left")
```

                                                Join

```python
#La diferencia de Join con Merge es que Join va directamente a los indices con los que se trabaja, y no a columnas.
```

```python
izq=pd.DataFrame({"A":["A0","A1","A2"],
             "B":["B0","B1","B2"]},
             index=["K0","K1","K2"])

der=pd.DataFrame({"C":["C0","C1","C2"],
             "D":["D0","D1","D2"]},
             index=["K0","K2","K3"])
izq
```

```python
der
```

```python
izq.join(der)
```

```python
#Con how="outer" se conservan los joins, pero se trae la informacion, 
#sin importar que este en la izquierda o dereche compartida.
izq.join(der,how="outer")
```

```python
#how="inner" junta solo la informacion comun
izq.join(der,how="inner")
```

```python
df2=pd.DataFrame({"A":["A4","A5","A6","A7"],
"B":["B4","B5","B6","B7"],
"C":["C4","C5","C6","C7"],
"D":["D4","D5","D6","D7"]})
df2
```

```python
#Normalmente las concatenaciones se hacen por el eje 0 (filas)
#ignore_index reorganiza los indices
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

```python
df4=pd.concat([df1,df2],axis=1)
df4
```

Merge

```python
#merge va con columnas, y no con indices, como join
izq=pd.DataFrame({"key":["k0","k1","k2","k3"],
             "A":["A0","A1","A2","A3"],
             "B":["B0","B1","B2","B3"]})
izq
```

```python
der=pd.DataFrame({"key":["k0","k1","k2","k3"],
             "C":["C0","C1","C2","C3"],
             "D":["D0","D1","D2","D3"]})
der
```

```python
izq.merge(der,on="key")
```

```python
center=pd.concat([izq,der],ignore_index=True)
center
```

```python
izq=pd.DataFrame({"key":["k0","k1","k2","k3"],
             "A":["A0","A1","A2","A3"],
             "B":["B0","B1","B2","B3"]})

der=pd.DataFrame({"key_2":["k0","k1","k2","k3"],
             "C":["C0","C1","C2","C3"],
             "D":["D0","D1","D2","D3"]})

izq.merge(der,left_on="key",right_on="key_2")
```

```python
izq=pd.DataFrame({"key":["k0","k1","k2","k3"],
             "A":["A0","A1","A2","A3"],
             "B":["B0","B1","B2","B3"]})

der=pd.DataFrame({"key_2":["k0","k1","k2",np.nan],
             "C":["C0","C1","C2","C3"],
             "D":["D0","D1","D2","D3"]})

izq.merge(der,left_on="key",right_on="key_2",how="right")
```

Join

```python
#join va directamente a los indices, y no a columnas especificas.
izq=pd.DataFrame({"A":["A0","A1","A2"],
                  "B":["B0","B1","B2"]},
                  index=["k0","k1","k2"])

der=pd.DataFrame({"C":["C0","C1","C2"],
                  "D":["D0","D1","D2"]},
                  index=["k0","k2","k3"])

izq.join(der,how="outer")
```
